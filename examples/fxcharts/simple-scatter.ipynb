```java
import java.util.Random;
import java.util.stream.IntStream;
import java.util.Arrays;
```

```java
var r = new Random();
```

```java
var y = r.doubles(105, 5, 10).toArray();
var x = IntStream.range(2,102).toArray();
```

```java
import javafx.scene.chart.NumberAxis;
import javafx.scene.chart.ScatterChart;
import javafx.scene.chart.XYChart;
```

```java
var xAxis = new NumberAxis();
xAxis.setLabel("X Data");

var yAxis = new NumberAxis();
yAxis.setLabel("Random Y Data");

var scatterChart = new ScatterChart(xAxis, yAxis);
```

```java
var series = new XYChart.Series();
series.setName("Random Data");

var intStream = Arrays.stream(x);
var data = intStream.mapToObj(i -> new XYChart.Data<Integer,Double>(i, y[i])).toList();

series.getData().addAll(data);

scatterChart.getData().setAll(series);

display(scatterChart);
```
